#### Question 1: Opioid Epidemic in VA

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_op = pd.read_csv("OpiodsVA.csv")     # Read in data from CSV to pandas dataframe
fig, ax = plt.subplots()

ax.scatter(df_op["MedianHouseholdIncome"], df_op["FPOO-Rate"], alpha=.5, c="b")   # Create scatterplot
fig.suptitle("Opioid Deaths and Median Income by County in VA")   # Set title and x,y axis labels
ax.xaxis.set_label_text("Median Household Income")
ax.yaxis.set_label_text("Opioid Overdoses per 100k")
fig.set_size_inches(7,5)
plt.show()
```

The scatterplot above demonstrates a moderate negative correlation between opioid overdoses and median household income in VA. Scatterplots are the most effective graph type to visualize how the relationship between two variables changes together because we can easily identify trends. 

```python
fig, ax = plt.subplots()

ax.scatter(df_op["FPOO-Rate"], df_op["FFHO-Rate"], alpha=.5, c="r")   # Create scatterplot
fig.suptitle("Opioid vs. Heroin Overdoses by County in VA")   # Set title and x,y axis labels
ax.xaxis.set_label_text("Opioid Overdoses per 100k")
ax.yaxis.set_label_text("Heroin Overdoses per 100k")
fig.set_size_inches(7,5)
plt.show()
```

The relationship between opioid and heroin overdoes in Virginia appears to have a slight positive correlation, but not as strong as income. Once again, scatterplots are the most effective visualization for visualizaing relationships between two variables.

#### Question 2: US Presidents Birthplace

```python
df_prez = pd.read_csv("presidents.csv")     # Read in data from .csv to data frame

states_dic = dict(df_prez["State"].value_counts())     # Count frequency of each state in dictionary
x = list(states_dic.keys())     # Extract keys as list and assign to x
y = list(states_dic.values())   # Extract values as list and assign to y

states = x[0:7]    # Extract top 7 values for graph
freq = y[0:7]

fig, ax = plt.subplots()
ax.bar(states, freq, color="g", alpha = .5)

fig.suptitle("Most Common Birth States of US Presidents")
fig.set_size_inches(8,6)

ax.xaxis.set_label_text("Birth State")
ax.yaxis.set_label_text("Frequency")
plt.show()
```

Virginia is the birth state to the most US presidents with 8, followed by Ohio (7), New York (5), and Massachusetts (4). The bar graph is the most effective method to show this data in order to quickly compare the frequencies side-by-side and quickly shows the top values.

#### Question 3: US Vehicle Sales

```python
df_cars = pd.read_csv("TOTALNSA.csv")

# Plot Data as Time Series line graph
fig, ax = plt.subplots()
ax.plot(df_cars["DATE"], df_cars["TOTALNSA"])

# Set title, axis labels, and size
fig.suptitle("US Car Sales from 1976-2018")
fig.set_size_inches(8,6)

ax.xaxis.set_label_text("Year")
ax.yaxis.set_label_text("Total Monthly Sales")

ax.xaxis.set(ticks=range(1,516, 96), ticklabels=["1976", "1984", "1992", "2000", "2008","2016"])   # Manually set ticks

plt.show()

```

The graph above shows that car sales have been generally increasing over the time period, with occasional declines likely caused by economic recessions. The time series is best represented by a line graph because the reader can quickly identify trends, seasonality, and cyclicality over the indicated time period.
